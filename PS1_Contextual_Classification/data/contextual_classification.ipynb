# Spectroscopic transients

```python
client = MongoClient('localhost:27017')
db = client.detection_context_features_spec
```

```python
projection = {"_id" : 0, "transient_object_id": 1, "class" : 1, \
              "sdss_host_offset" : 1, "has_qso_colours" : 1, "is_star" : 1, \
              "is_gal" : 1, "sdss_host" : 1, "r" : 1, "survey" : 1, \
              "photoz" : 1, "err_photoz" : 1, "ra" : 1, "dec" : 1}
```

```python
"""
query = {"$and": [{"host_r":{"$gte":20}},\
                  {"host_r":{"$ne":"NULL"}},\
                  {"class":{"$ne":"NULL"}},\
                  {"class":{"$ne":"Other"}},\
                  {"ra":{"$ne":None}},\
                  {"dec":{"$ne":None}},\
                  {"sdss_host_offset":{"$ne":None}}]}
"""
query = {"$and": [{"r":{"$ne":"NULL"}},\
                  {"class":{"$ne":"NULL"}},\
                  {"class":{"$ne":"Other"}},\
                  {"class":{"$ne":None}},\
                  {"ra":{"$ne":None}},\
                  {"dec":{"$ne":None}},\
                  {"sdss_host_offset":{"$ne":None}}]}
```

```python
num_results = db.detection_context_features_spec.find(query, projection).count()
results = db.detection_context_features_spec.find(query, projection)
```

```python
print num_results
```

```python
X_train = np.ones((num_results, 8)) # 8 features
y_train = np.ones((num_results,))
ids_train = []
surveys_train = []
for i,result in enumerate(results):
    #print i,result["survey"], result["ra"]
    if result["sdss_host_offset"] != "NULL" :
        X_train[i,0] *= result["sdss_host_offset"]
        X_train[i,1] *= result["r"]
        X_train[i,2] *= result["is_gal"]
        X_train[i,3] *= result["has_qso_colours"]
        X_train[i,6] *= float(result["ra"])
        X_train[i,7] *= float(result["dec"])
        try:
            X_train[i,4] *= result["photoz"]
            X_train[i,5] *= result["err_photoz"]
        except ValueError: 
            X_train[i,4] *= 0
            X_train[i,5] *= 0
        ids_train.append(result["transient_object_id"])
        surveys_train.append(result["survey"])
        if result["class"] == "SN":
            y_train[i] *= 1
        elif result["class"] == "AGN":
            y_train[i] *= 0
        elif result["class"] == "STAR":
            y_train[i] *= 2
        else:
            print "[!] " + str(result)
    else:
        print "[!] " + str(result)
```

```python
from sklearn import svm
from sklearn import preprocessing
from sklearn.decomposition import PCA
```

```python
#scaler_1 = preprocessing.StandardScaler().fit(X_train[:,:6])
#X_scaled_train = scaler_1.transform(X_train[:,:6])
#X_normalized_train = preprocessing.normalize(X_scaled_train, norm='l2')
X_normalized_train = preprocessing.normalize(X_train[:,:4], norm='l2')
pca = PCA(n_components=2)
X_r_train = pca.fit(X_normalized_train).transform(X_normalized_train)
scaler_1 = preprocessing.StandardScaler().fit(X_r_train)
X_r_train = scaler_1.transform(X_r_train)
```

```python
plt.scatter(X_r_train[np.where(y_train==1)[0],0], X_r_train[np.where(y_train==1)[0],1], color="#66FF33", label="SN")
plt.scatter(X_r_train[np.where(y_train==0)[0],0], X_r_train[np.where(y_train==0)[0],1], color="#3366FF", label="AGN")
plt.scatter(X_r_train[np.where(y_train==2)[0],0], X_r_train[np.where(y_train==2)[0],1], color="#FF0066", label="STAR")
```

```python
y_train[np.where(y_train==2)[0]] -= 2
clf = svm.SVC(kernel='rbf', C=100, gamma=100, probability=True).fit(X_r_train, y_train)
```

```python
h = .02
x_min, x_max = X_r_train[:, 0].min() - .1, X_r_train[:, 0].max() + .1
y_min, y_max = X_r_train[:, 1].min() - .1, X_r_train[:, 1].max() + .1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                     np.arange(y_min, y_max, h))
```

```python
Z = clf.predict_proba(np.c_[xx.ravel(), yy.ravel()])[:,1]
Z = Z.reshape(xx.shape)

plt.contourf(xx, yy, Z, levels=np.linspace(Z.min(), Z.max(), 11),
                 cmap=plt.cm.Purples)

plt.contour(xx, yy, Z, levels=[0.9], linewidths=2, colors='white')
plt.scatter(X_r_train[np.where(y_train==1)[0],0], X_r_train[np.where(y_train==1)[0],1], color="#66FF33", label="SN", alpha=.5)
plt.scatter(X_r_train[np.where(y_train==0)[0],0], X_r_train[np.where(y_train==0)[0],1], color="#3366FF", label="AGN", alpha=.5)
plt.xlim(xmax=-0.5)
```

```python
from sklearn.metrics import roc_curve

pred_train = clf.predict_proba(X_r_train)[:,1]

fpr, tpr, thresholds = roc_curve(y_train, pred_train)

plt.plot(1-tpr, fpr, "k", lw=5)
plt.plot(1-tpr, fpr, "#FF0066", lw=4)
plt.xlabel("MDR")
plt.ylabel("FPR")
```
